# 가상화폐 시세 판단하기

> 최고가와 최저가의 차이를 변동폭으로 정의할 때 (시가 + 변동폭)이 최고가 보다 높을 경우 "상승장", 그렇지 않은 경우 "하락장" 문자열을 출력하라.

|Key Name        |     Description|
|---|---|
|opeing_price    |최근 24시간 내 시작 거래금액|
|closing_price   |최근 24시간 내 마지막 거래금액|
|min_price       |최근 24시간 내 최저 거래금액|
|max_price       |최근 24시간 내 최고 거래금액|

````python
import requests
import pprint
btc = requests.get('https://api.bithumb.com/public/ticker/btc').json()['data']
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(btc)
````

````python
# 아래에 코드를 작성하세요.
max_price = float(btc["max_price"])

rangePrice = float(btc["max_price"]) - float(btc["min_price"])
opening_price = float(btc["opening_price"])
sum_price = rangePrice + opening_price

if sum_price > max_price:
        print("상승장")
else:
        print("하락장")
````

## 모든 가상화폐 시세 판단하기


> currrency dictionary에 담겨있는 모든 가상화폐 거래내역을 바탕으로 출력하라.
>
> 예외 처리에 유의하세요.



```
예시 출력)
BTC 상승장
ETH 상승장
DASH 상승장
LTC 하락장
ETC 하락장
XRP 상승장
BCH 상승장
...
```


````python
currency = requests.get('https://api.bithumb.com/public/ticker/all').json()['data']
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(currency)
````

````python
# 아래에 코드를 작성하세요.
for i in currency:
    if type(currency[i]) != dict:
        continue
    print(i, end = " ")
    max_price = float(currency[i]["max_price"])

    rangePrice = float(currency[i]["max_price"]) - float(currency[i]["min_price"])
    opening_price = float(currency[i]["opening_price"])
    sum_price = rangePrice + opening_price

    if sum_price > max_price:
        print("상승장")
    else:
        print("하락장")
````

````python
# 아래에 코드를 작성하세요.
for i, v in currency.items():
    if type(currency[i]) != dict:
        continue
    max_price = float(v["max_price"])

    rangePrice = float(v["max_price"]) - float(v["min_price"])
    opening_price = float(v["opening_price"])
    sum_price = rangePrice + opening_price

    if sum_price > max_price:
        print(f"{i:5}상승장")
    else:
        print(f"{i:5}하락장")
````
